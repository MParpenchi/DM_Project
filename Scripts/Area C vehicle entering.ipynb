```python
import pandas as pd
import glob
import os

# 1. Dosyaların olduğu klasör yolu
path = r"C:\Users\User\Desktop\Data Management\2023 Area C entering"
all_files = glob.glob(os.path.join(path, "ingressi_areac_2023_*.csv"))

# 2. Dosyaları okuyup bu listeye kaydedeceğiz
data_list = []

print("Dosyalar okunuyor...")

for filename in all_files:
    # Her bir dosyayı okuyup listeye ekle
    df = pd.read_csv(filename, sep=',', low_memory=False)
    data_list.append(df)
    print(f"{os.path.basename(filename)} listeye eklendi.")

print("\nİşlem tamam! Toplam", len(data_list), "adet ay verisi listede.")
```

```python
import pandas as pd

# Sonuçları toplamak için bir liste oluşturuyoruz
monthly_summary = []

print("Aylık toplamlar hesaplanıyor, lütfen bekleyin...")

for df in data_list:
    # 1. 'dataora' sütununu tarihe çevir ve saat dilimini kaldır (Hata almamak için)
    df['dataora'] = pd.to_datetime(df['dataora'], errors='coerce').dt.tz_localize(None)
    
    # 2. 'numero_transiti' sütununu sayısal formata zorla (Hatalı verileri temizler)
    df['numero_transiti'] = pd.to_numeric(df['numero_transiti'], errors='coerce')
    
    # 3. Boş (geçersiz) tarihli satırları temizle
    df = df.dropna(subset=['dataora'])
    
    # 4. Ay bilgisini belirle (Dosyadaki ilk satırdan YYYY-MM formatında al)
    month_label = df['dataora'].dt.strftime('%Y-%m').iloc[0]
    
    # 5. O ayın toplam araç geçiş sayısını hesapla
    total_vehicles = df['numero_transiti'].sum()
    
    # Listeye ekle
    monthly_summary.append({
        'Month': month_label,
        'Total_Traffic': int(total_vehicles)
    })

# 6. Tabloyu oluştur ve ayları (Ocak'tan Aralık'a) sıraya diz
final_df = pd.DataFrame(monthly_summary).sort_values(by='Month').reset_index(drop=True)

# Çıktıyı göster
print("\n--- 2023 MILANO AREA C AYLIK TOPLAM GİRİŞLER ---")
print(final_df)
```
